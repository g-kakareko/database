```python
import pandas as pd
import numpy as np

from datetime import datetime
import random
```

```python
data_csv = '/home/grzegorz/Desktop/Database/Project/data.csv' 
df_names = pd.read_csv(data_csv, sep=',')

club_csv = '/home/grzegorz/Desktop/Database/Project/clubs.csv' 
df_clubs = pd.read_csv(club_csv, sep=',')
```

```python
df_clubs
```

```python
trainer_header = ['Name', 'DOB', 'Club', 'trainer_id']
df_trainer = pd.DataFrame(columns=trainer_header)

# trainer_header = ['Club', 'Location', 'Capacity']
# df_clubs = pd.DataFrame(columns=trainer_header)
```

```python
df_trainer['Name'] = df_names['Trainer'].iloc[0:50]
df_trainer['DOB'] = df_names['Trainer_DOB'].iloc[0:50]
tr = np.random.randint(len(df_clubs), size=len(df_trainer))
df_trainer['Club'] = df_clubs['Club'].iloc[tr].values
df_trainer['trainer_id'] = np.arange(len(df_trainer))
df_trainer
```

```python
# np.random.randint(len(df_clubs), size=len(df_trainer))
df_clubs['stadium_id'] = np.arange(len(df_clubs))
df_clubs
```

```python
runner_header = ['Name', 'DOB', 'Club', 'gender','trener_id']
# Male
df_runner_M = pd.DataFrame(columns=runner_header)
df_runner_M['Name'] = df_names['Runner_male'].iloc[0:300]
df_runner_M['DOB'] = df_names['Runner_male_DOB'].iloc[0:300]
tr = np.random.randint(len(df_clubs), size=len(df_runner_M))
df_runner_M['Club'] = df_clubs['Club'].iloc[tr].values
tr = np.random.randint(len(df_trainer), size=len(df_runner_M))
df_runner_M['trener_id'] = df_trainer['trainer_id'].iloc[tr].values
df_runner_M['gender'] = np.array(['M']*(len(df_runner_M)))
```

```python
# Female
df_runner_F = pd.DataFrame(columns=runner_header)
df_runner_F['Name'] = df_names['Runner_female'].iloc[0:300]
df_runner_F['DOB'] = df_names['Runner_male_DOB.1'].iloc[0:300]
tr = np.random.randint(len(df_clubs), size=len(df_runner_F))
df_runner_F['Club'] = df_clubs['Club'].iloc[tr].values
tr = np.random.randint(len(df_trainer), size=len(df_runner_F))
df_runner_F['trener_id'] = df_trainer['trainer_id'].iloc[tr].values
df_runner_F['gender'] = np.array(['F']*(len(df_runner_F)))
```

```python
# print(df_runner_M)
df_runner = pd.concat([df_runner_M, df_runner_F])
```

```python
# Building the Competition table
competition_header = ['stadium_id', 'date', 'attendence', 'weather']
df_competition = pd.DataFrame(columns=competition_header)
# we generate 100 100 comeptition
# weather types:
# Rainy- R
# Sunny- S
# Cloudy- K
# Hot- H
# Cold- C
# Dry- D
# Wet- W
weather_types = ['R', 'S', 'K', 'H', 'C', 'D', 'W']
wt = list()
date_list = list()
att_list = list()
for i in np.arange(1,100): 
    n = random.randint(0, (len(weather_types)-1))
    wt.append(weather_types[n])
    year = random.randint(2015, 2018)
    month = random.randint(4, 8)
    day = random.randint(1, 30)
    comp_date = datetime(year, month, day)
    date_list.append(comp_date)
    att_list.append(random.randint(1000, 3000))
    
    
df_competition['date'] = date_list
df_competition['attendence'] = att_list
df_competition['weather'] = wt
tr = np.random.randint(len(df_clubs), size=len(df_competition))
df_competition['stadium_id'] = df_clubs['stadium_id'].iloc[tr].values
df_competition['competition_id'] = np.arange(len(df_competition))
```

```python
# Generate the Results Table
result_header = ['Name', 'DOB', 'Club', 'gender', 'result', 'discipline', 'date', 'competition_id','result_id']
df_result = pd.DataFrame(columns=result_header)
discip_head = np.array(['100', '200', '400', '800', '1000', '1500'])
n = 100000
tr = np.random.randint(len(df_runner), size=n)
# for i in np.arange(1,100): 
#     n = random.randint(0, (len(weather_types)-1))
df_result['Name'] = df_runner['Name'].iloc[tr].values
df_result['DOB'] = df_runner['DOB'].iloc[tr].values
df_result['Club'] = df_runner['Club'].iloc[tr].values
df_result['gender'] = df_runner['gender'].iloc[tr].values
tr = np.random.randint(len(discip_head), size=n)
df_result['discipline'] = discip_head[tr]

tr = np.random.randint(len(df_competition), size=n)
df_result['date'] = df_competition['date'].iloc[tr].values
df_result['competition_id'] = df_competition['competition_id'].iloc[tr].values
df_result['result_id'] = np.arange(len(df_result))

results = list()
for index, row in df_result.iterrows():
    if row['gender']=='M':
        if row['discipline']=='100':
            res = np.random.randint(10,13) + np.random.randint(0,100)/100
        elif row['discipline']=='200':
            res = np.random.randint(20,23) + np.random.randint(0,100)/100
        elif row['discipline']=='400':
            res = np.random.randint(45,51) + np.random.randint(0,100)/100
        elif row['discipline']=='800':
            res = np.random.randint(105,115) + np.random.randint(0,100)/100
        elif row['discipline']=='1000':
            res = np.random.randint(140,160) + np.random.randint(0,100)/100
        else:
            res = np.random.randint(210,240) + np.random.randint(0,100)/100
    else:
        if row['discipline']=='100':
            res = np.random.randint(12,15) + np.random.randint(0,100)/100
        elif row['discipline']=='200':
            res = np.random.randint(23,26) + np.random.randint(0,100)/100
        elif row['discipline']=='400':
            res = np.random.randint(51,57) + np.random.randint(0,100)/100
        elif row['discipline']=='800':
            res = np.random.randint(125,140) + np.random.randint(0,100)/100
        elif row['discipline']=='1000':
            res = np.random.randint(165,195) + np.random.randint(0,100)/100
        else:
            res = np.random.randint(245,270) + np.random.randint(0,100)/100
    # Female
    results.append(res)
df_result['result'] = results  
```

```python
df_result_60 = df_result[df_result['competition_id']==60]
df_result = df_result[df_result['competition_id']!=60]
df_result_61 = df_result[df_result['competition_id']==61]
df_result = df_result[df_result['competition_id']!=61]
```
